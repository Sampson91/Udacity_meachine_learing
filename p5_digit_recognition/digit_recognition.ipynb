### 练习：合成数据

你需要随机取随机张图片，然后将它们拼接成新的图片。

你需要设置20%的数据作为验证集，以保证模型没有过拟合。

```python
import numpy as np
from sklearn.model_selection import train_test_split
from itertools import chain
n_class, n_len, width, height = 11, 5, 28, 28

def generate_dataset(X, y):
    X_len = X.shape[0]
    X_gen = np.zeros((X_len, height, width * n_len, 1), dtype=np.uint8)
    y_gen = [np.zeros((X_len, n_class), dtype=np.uint8) for i in range(n_len)]
    # TODO: 随机取1~5个数字，并拼接成新的图片
    #     y_gen = np.array([[y[j] for j in random.sample(range(0,X_len), 5)] for i in range(0,X_len)])
    for i in range(0, X_len):
        rand = random.randint(1, 5)
        for j in range(rand):
            r = random.randint(0,X_len-1)
            y_gen[j][i][y[r]] = 1                         
            X_gen[i][:, j * width:(j + 1) * width,0] = X[r]
        for k in range(rand,5):
            y_gen[k][i][10] = 1                      
#             X_gen[i][:, k * width:(k + 1) * width,0] = 0
    return X_gen, y_gen



X_raw_train, X_raw_valid, y_raw_train, y_raw_valid = train_test_split(X_raw,y_raw,test_size=0.2,random_state = 22)

X_train, y_train = generate_dataset(X_raw_train, y_raw_train)
X_valid, y_valid = generate_dataset(X_raw_valid, y_raw_valid)
X_test, y_test = generate_dataset(X_raw_test, y_raw_test)
```

### 练习：设计并测试一个模型架构

设计并实现一个能够识别数字序列的深度学习模型。为了产生用于测试的合成数字序列，你可以进行如下的设置：比如，你可以限制一个数据序列最多五个数字，并在你的深度网络上使用五个分类器。同时，你有必要准备一个额外的“空白”的字符，以处理相对较短的数字序列。

在使用 Keras 搭建模型的时候，你可以使用 [泛型模型](http://keras-cn.readthedocs.io/en/latest/models/model/) 的方式来搭建多输出模型。

```python
from keras.models import Model
from keras.layers import *
from keras.models import Sequential

# TODO: 构建你的模型
batch_size = 521
epochs = 12
num_classes = 11
img_rows, img_cols = 28, 140

# x_train = X_train.reshape(X_train.shape[0], img_rows, img_cols, 1)
# x_test = X_test.reshape(X_test.shape[0], img_rows, img_cols, 1)
input_shape = (img_rows, img_cols, 1)

x_train = X_train.astype('float32')
x_test = X_test.astype('float32')
x_train /= 255
x_test /= 255
print('x_train shape:', x_train.shape)
print(x_train.shape[0], 'train samples')
print(x_test.shape[0], 'test samples')
```

```python
from keras import regularizers

input_data = Input(shape=input_shape)

x = Conv2D(32,(3,3),activation='relu')(input_data)
x = MaxPool2D(pool_size=(2,2))(x)
x = Conv2D(64,(3,3),activation='relu')(x)
x = MaxPool2D(pool_size=(2,2))(x)
x = Conv2D(128,(3,3),activation='relu')(x)
x = MaxPool2D(pool_size=(2,2))(x)
# x = Conv2D(64,(3,3),activation='relu')(x)
# x = MaxPool2D(pool_size=(2,2))(x)

x = Flatten()(x)
x = Dropout(0.5)(x)
# x = Dense(256,activation='relu')(x)

y_out = [Dense(11, activation='softmax')(x) for i in range(n_len)]
model = Model(inputs=input_data,outputs=y_out)

from keras.losses import categorical_crossentropy
from keras.optimizers import Adadelta
model.compile(loss=categorical_crossentropy,
              optimizer=Adadelta(),
              metrics=['accuracy'])

```

### 问题 3
_你最终的模型架构是什么样的？（什么类型的模型，层数，大小, 如何连接等）
**回答：**

    该模型采用keras搭建的10层卷积升神经网络，第一层为input输入层对（28,140,1）的图像进行输入，（2,4,6）层为（32,64,128）个核的尺寸为3×3的卷积层，(3,5,7）为2×2的池化层，第八层为展开层，把数据转换为一维，第九层为dropout层。第十层为5个结果的输出层。
    
    其中网络中各个层的技术实现方式为：
        1.在网络中卷积层采用3×3像素的尺寸的窗口在图像中进行滑动对每个3×3的窗口根据学习的权重进行加权求和，最后返还一个经过卷积的图片;获得的图片通过relu函数f(x)=max(0,x)来将所有负值像素替换为0。
        2.由于卷积层获取的图片尺寸没有发生变化，而为了减少模型的复杂度，通过压缩图片来实现即通过神经网络中下采样来稀疏参数，该模型采用最大值池化Max_pool方法通过2×2的像素中选择最大的值作为下一层传输。。
        3.Dropout层是指在模型训练时随机概率让网络某些隐含层节点的权重不工作，不工作的那些节点可以暂时认为不是网络结构的一部分，但是它的权重得保留下来（只是暂时不更新而已），因为下次样本输入时它可能又得工作了。
        4.对构建好的模型通过梯度下降的方法学习卷积核的权重，对于梯度下降中各个权重参数有些维度变化快，有些维度变化慢；有些维度是负的斜坡，有些维度是正的斜坡采用相同的Learning Rate是不科学的，比如有的参数可能已经到了仅需要微调的阶段，但又有些参数由于对应样本少等原因，还需要较大幅度的调动。理想的方案是根据参数每个维度的变化率，采用对应的Learning Rate。Adadelta优化方法为每个参数计算自适应的Learning Rate，给不经常更新的参数以较大的Learning Rate， 给经常更新的参数以较小的Learning Rate并且Adadelta与其它优化算法相比解决了1）学习率是单调递减的，训练后期学习率非常小。2）其需要手工设置一个全局的初始学习率。3）更新W时，左右两边的单位不同问题。
        5.Softmax回归模型，该模型是logistic回归模型在多分类问题上的推广，在多分类问题中，类标签y可以取两个以上的值。 Softmax回归模型对于诸如MNIST手写数字分类等问题是很有用的，该问题的目的是辨识10个不同的单个数字对于给定的测试输入 x，我们想用假设函数针对每一个类别j估算出概率值 p(y=j | x)。也就是说，我们想估计 x 的每一种分类结果出现的概率。因此，我们的假设函数将要输出一个k 维的向量（向量元素的和为1）来表示这k个估计的概率值，也就是softmax把一个k维的real value向量（a1,a2,a3,a4….）映射成一个（b1,b2,b3,b4….）其中bi是一个0-1的常数，然后可以根据bi的大小来进行多分类的任务，如取权重最大的一维。

### 练习：计算你的模型准确率

我们刚才得到了模型每个数字的准确率，现在让我们来计算整体准确率，按照完全预测正确数字序列的标准来计算。

比如 1,2,3,10,10 预测成了 1,2,10,10,10 算错，而不是算对了80%。

```python
def evaluate(model,X_test, y_test):
    # TODO: 按照错一个就算错的规则计算准确率
    y_test_predict = model.predict(X_test)
    y_test_predict = np.array(y_test_predict)
    y_test_predict = np.argmax(y_test_predict, axis = 2)
    y_test = np.argmax(y_test,axis = 2)
    y_test_predict = y_test_predict.T
    y_test = y_test.T
    number = 0
    for i in range(len(y_test_predict)):
        if (y_test_predict[i] == y_test[i]).all():
            number += 1
    accuracy = float(number)/len(X_test)
    return accuracy
evaluate(model,X_valid, y_valid )
```
